```python
import json
import os
from datetime import datetime
```

```python
def loadJson(fileName):
    """
        Load Json file
    """
    if os.path.exists(f'{fileName}.json'):
        if os.stat(f"{fileName}.json").st_size == 0:
            return {}
        else:
            with open(f'{fileName}.json','r') as f:
                record=json.loads(f.read())
            return record    
    else:
        return {}
```

```python
def dumpsJson(fileName,r):
    """
        This method is used for save progress into file
    """
    if input(f"Do you want to save all {fileName} updates? Y/N  ").lower()=='y':
        
        txt=json.dumps(r)
        with open(f'{fileName}.json','w') as f:
          f.write(txt)
        print('***************  Record Saved  ***************')
    else:
        print("*************** Record Not saved ***************")
```

```python
def displayRecord(r):
    """
        This function is used for displaying records.
            1. Display Adding and Updating Record
            2. Display Sells Record
    """
    if r!={}:
        print("*************** Record ***************")
        for i in r:
            print(i,":=>",r[i])
    else:
        print("*************** Record is Empty ***************")
```
